```python
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C


# load data in numpy objects
f_initial_inputs =np.load("initial_data/function_" + fn + "/initial_inputs.npy")
f_initial_outputs =np.load("initial_data/function_" + fn + "/initial_outputs.npy")


X = np.append(f_initial_inputs, [Q1], axis=0)
X = np.append(X, [Q2],  axis=0)
X = np.append(X, [Q3],  axis=0)
X = np.append(X, [Q4],  axis=0)
X = np.append(X, [Q5],  axis=0)
X = np.append(X, [Q6],  axis=0)
X = np.append(X, [Q7],  axis=0)
X = np.append(X, [Q8],  axis=0)
X = np.append(X, [Q9],  axis=0)
X = np.append(X, [Q10], axis=0)
X = np.append(X, [Q11], axis=0)
X = np.append(X, [Q12], axis=0)


Y = np.append(f_initial_outputs, [Obs1], axis=0)
Y = np.append(Y, [Obs2],  axis=0)
Y = np.append(Y, [Obs3],  axis=0)
Y = np.append(Y, [Obs4],  axis=0)
Y = np.append(Y, [Obs5],  axis=0)
Y = np.append(Y, [Obs6],  axis=0)
Y = np.append(Y, [Obs7],  axis=0)
Y = np.append(Y, [Obs8],  axis=0)
Y = np.append(Y, [Obs9],  axis=0)
Y = np.append(Y, [Obs10], axis=0)
Y = np.append(Y, [Obs11], axis=0)
Y = np.append(Y, [Obs12], axis=0)

print("Function Number: ", functions_dict[fn])
print("Lenght of X: ", len(X))
print("Lenght of Y: ", len(Y))


# Define the kernel and GaussianProcessRegressor
# kernel = C(1.0) * RBF(length_scale=1.0)
kernel = RBF(length_scale=0.1, length_scale_bounds='fixed')
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)

# Fit the model
gp.fit(X, Y)

# Generate next X values
X_next = np.random.rand(50, functions_dict[fn])  # Generating 50 random points in the same space

Y_pred, sigma = gp.predict(X_next, return_std=True)

# Print the predicted values
print()
print("Next potentially best X values:")
formatted_array = '\n'.join(['-'.join([f"{num:.6f}" for num in row]) for row in X_next])

print(formatted_array)

#The next best X values can be chosen based on the highest predicted Y values or the highest uncertainty for exploration.
```

```python
print("\nPredicted Y values:")
print(Y_pred)

print("\nIf choose to exploit then look for Y Max")
print("Y_pred_max = ", Y_pred.max())
print("X value for the Y_pred_max = ",  X_next[np.argmax(Y_pred)])

print("\nPrediction uncertainty:")
print(sigma)
print("\nIf choose exploration then look for Sigma Max")
print("sigma max = ", sigma.max())
print("X value for the sigma_max = ",  X_next[np.argmax(sigma)])
```
